```python
def boxplotter(series):
    if series.dtype == 'O' or len(series.unique()) < 10:
        plt.figure(figsize=(20,10));
        plt.title(f'{series.name} vs. Sale Price');
        sns.boxplot(x=series.name, y='SalePrice', data=housing);
    else:
        print('So many columns.  Try something else.')
```

```python
def violinplotter(series):
    if series.dtype == 'O' or len(series.unique()) < 10:
        plt.figure(figsize=(20,10));
        plt.title(f'{series.name} vs. Sale Price');
        sns.violinplot(x=series.name, y='SalePrice', data=housing);
    else:
        print('So many columns.  Try something else.')
```

```python
def splitplots(series, explanationSrs = housing.Gr_Liv_Area):
    column = series.name
    for value in housing[column].unique():
        plt.figure();
        sns.regplot(x=explanationSrs.name, y='SalePrice', data=housing[housing[column] == value], scatter_kws={'alpha':0.4});
        plt.title(f'{column} : {value}');
```

```python
terms = reduce((lambda x,y: x+y),[column + ' + ' for column in housing.drop('SalePrice', axis=1).columns])[:-3]
```

```python
terms=reduce((lambda x,y: x + " + " + y), ['Gr_Liv_Area', 'Exter_Qual', 'Overall_Qual', 'Year_Built', 'Bsmt_Qual', 'Garage_Cars', 'MS_SubClass', 'Kitchen_Qual', 'Neighborhood', 'Bsmt_Full_Bath', 'Overall_Cond'])
```

```python
areas = []

for column in housing.columns:
    if 'SF' in column or 'area' in column.lower() and column != 'Gr_Liv_Area':
        areas.append(column)
```

```python
terms = reduce((lambda x,y: x+y),[column + ' + ' for column in housing\
                                  .drop(['SalePrice', \
                                         'Lot_Frontage', 'First_Flr_SF'
                                        ] + areas, axis=1).columns]\
              )[:-3]
```

```python
interactions = ' + Lot_Area:Lot_Shape:Land_Contour + Overall_Qual:Gr_Liv_Area + Overall_Qual:Gr_Liv_Area:Sale_Type - 1'
formula = f'SalePrice ~ {terms}'
y, x = patsy.dmatrices(formula, housing)
```

```python
lm = LinearRegression()
lasso = LassoCV(n_alphas=100, cv=10)
ridge = RidgeCV(alphas=np.logspace(0,5,100), cv=10)
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.4, shuffle=True)

scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

```python
lasso.fit(xtrain, ytrain)
lasso.score(xtrain, ytrain) , lasso.score(xtest, ytest)
```

```python
mean_squared_error(ytest, lasso.predict(xtest))**0.5
```
